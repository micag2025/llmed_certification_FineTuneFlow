```python
# =====================================================
# Upload FINAL BART MERGED MODEL to Hugging Face Hub
# =====================================================

import os
from huggingface_hub import HfApi, HfFolder, create_repo

# -------------------------
# Config
# -------------------------
HF_USERNAME = "dlaima"      # <--- SET YOUR USERNAME
REPO_NAME = "bart-highlightsum-merged"

PUSH_MERGED = True          # âœ… Upload merged model
PUSH_LORA = False           # âŒ Do NOT upload LoRA adapters unless you want them

MERGED_PATH = "/content/llmed_certification_FineTuneFlow/ft_outputs/bart_merged_highlightsum"
LORA_PATH = "/content/llmed_certification_FineTuneFlow/ft_outputs/bart_lora_highlightsum"

# -------------------------
# Authenticate
# -------------------------
print("ðŸ”‘ Logging into Hugging Face...")
!huggingface-cli login

token = HfFolder.get_token()
if token is None:
    raise ValueError("âŒ No HF token found. Please log in.")

api = HfApi()

# -------------------------
# Create repo
# -------------------------
full_repo_name = f"{HF_USERNAME}/{REPO_NAME}"

try:
    create_repo(full_repo_name, exist_ok=True)
    print(f"ðŸ“ HF Repo ready: {full_repo_name}")
except Exception as e:
    print("âš ï¸ Repo exists:", e)

# -------------------------
# Upload merged model
# -------------------------
if PUSH_MERGED:
    print("\nðŸ“¤ Uploading MERGED model...")
    api.upload_folder(
        folder_path=MERGED_PATH,
        repo_id=full_repo_name,
        repo_type="model",
        commit_message="Upload merged BART HighlightSum model"
    )
    print("âœ… Merged model uploaded!")

# -------------------------
# (Optional) Upload LoRA adapter
# -------------------------
if PUSH_LORA:
    print("\nðŸ“¤ Uploading LoRA adapter...")
    api.upload_folder(
        folder_path=LORA_PATH,
        repo_id=full_repo_name,
        repo_type="model",
        commit_message="Upload LoRA adapter"
    )
    print("âœ… LoRA adapter uploaded!")

print("\nðŸŽ‰ Upload complete!")
print(f"ðŸ”— HF Model URL: https://huggingface.co/{full_repo_name}")
```
